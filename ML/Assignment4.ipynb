### Example
<br>
Given function $$ y = x^2 + 3 $$

We know the `derivative`of f(x) is $$ \frac{dy}{dx} = 2x $$

##### Gradient Descent

Gradient descent is a mathematical algorithm that finds the minimum value of a function by **iteratively** moving in the direction of the **steepest descent**.

![grad-descent](https://miro.medium.com/max/1400/1*tQTcGTLZqnI5rp3JYO_4NA.png)


```python
def f(x):
    return x**2 + 3

def df(x):
    return 2*x
```

```python
def grad_descent(f, df, alpha, epochs):
    x = np.random.rand()*20 - 10
    cache = [x]
    for epoch in range(epochs):
        dy_dx = df(x)
        x -= dy_dx*alpha
        cache.append(x)
    return x, cache
```

```python
x_min, _ = grad_descent(f, df, 0.1, 20)
print(f"Local minimum is at x = {x_min}")
```

### Relationship with Learning rate alpha

```python
fig = plt.figure(figsize=(15, 6))
x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='y = x^2 + 3')
plt.xlabel('x')
plt.ylabel('y')
x_min, cache = grad_descent(f, df, 0.1, 20)
for i in range(len(cache)-1):
    plt.scatter(cache[i], f(cache[i]), color='red', s=100, )
    plt.plot((cache[i], cache[i+1]), (f(cache[i]), f(cache[i+1])), color='red', linewidth=2)
plt.title('y = x^2 + 3')
plt.show()
```
